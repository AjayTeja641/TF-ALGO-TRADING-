```python
!pip install pandas_ta

import yfinance as yf
import pandas as pd
import numpy as np
import pandas_ta as ta
import matplotlib.pyplot as plt

# ==========================================
# PART 1: GET RAW DATA
# ==========================================
print("1. Downloading Maximum Nifty 50 History...")

# We request data from 2000. Yahoo will return the earliest available (approx 2007).
df = yf.download('^NSEI', start='2000-01-01', end=None, progress=False)
# I had used the data of NSE India as mentioned in the presentation.

# Fixing  the  column structure in new yfinance versions
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

print(f"\n✅ DATA DOWNLOADED:")
print(f"   - Start Date: {df.index[0].date()}")
print(f"   - End Date:   {df.index[-1].date()}")
print(f"   - Total Days: {len(df)}")
```

```python
# ==========================================
# PART 2: CALCULATE 10 INDICATORS
# ==========================================
print("\n2. Calculating 10 Advanced Parameters...")
# these are the key parameters we are t4esting the mofel on.

# --- GROUP A: VOLATILITY (Fear) ---
# 1. Rolling Volatility
df['Volatility'] = df['Close'].pct_change().rolling(20).std()
# 2. ATR (Average True Range)
df['ATR'] = ta.atr(df['High'], df['Low'], df['Close'], length=14)
# 3. Bollinger Band Width
bb = ta.bbands(df['Close'], length=20, std=2)
df['BB_Width'] = (bb['BBU_20_2.0_2.0'] - bb['BBL_20_2.0_2.0']) / bb['BBM_20_2.0_2.0']

# --- GROUP B: MOMENTUM (Speed) ---
# 4. RSI
df['RSI'] = ta.rsi(df['Close'], length=14)
# 5. ROC (Rate of Change)
df['ROC'] = ta.roc(df['Close'], length=10)
# 6. MACD
macd = ta.macd(df['Close'])
df['MACD'] = macd['MACD_12_26_9']
# 7. Williams %R
df['WillR'] = ta.willr(df['High'], df['Low'], df['Close'], length=14)

# --- GROUP C: TREND (Direction) ---
# 8. ADX (Trend Strength)
adx = ta.adx(df['High'], df['Low'], df['Close'], length=14)
df['ADX'] = adx['ADX_14']
# 9. Distance from SMA 50
df['Dist_SMA50'] = (df['Close'] / ta.sma(df['Close'], length=50)) - 1
# 10. Distance from SMA 200
df['Dist_SMA200'] = (df['Close'] / ta.sma(df['Close'], length=200)) - 1

# Drop the first 200 rows (because SMA200 needs 200 days of past data)
df.dropna(inplace=True)

print(f"✅ Features Calculated. Usable Training Data: {len(df)} days.")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# ==========================================
# PART 3: TRAINING THE MODEL
# ==========================================
print("\n3. Training AI Model on Full History...")

# 1. Creating Labels (The Truth)

df['Future_Ret'] = df['Close'].shift(-20) / df['Close'] - 1

def label_regime(x):
    if x > 0.025: return 2   # BULL
    elif x < -0.025: return 0 # BEAR
    else: return 1            # SIDEWAYS

df['Target'] = df['Future_Ret'].apply(label_regime)

# Drop the last 20 rows (we don't know their future yet)
df_model = df.iloc[:-20].copy()

# 2. Setup Inputs (X) and Output (y)
features = ['Volatility', 'ATR', 'BB_Width', 'RSI', 'ROC',
            'MACD', 'WillR', 'ADX', 'Dist_SMA50', 'Dist_SMA200']

X = df_model[features]
y = df_model['Target']

# 3. Split Data (80% Train, 20% Test)
# Shuffle=False is CRITICAL for financial time-series
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# 4. Train the Random Forest
# We use 300 Trees to handle the large dataset
rf_model = RandomForestClassifier(n_estimators=300, min_samples_leaf=5, random_state=42)
rf_model.fit(X_train, y_train)

# 5. Show Results
preds = rf_model.predict(X_test)
print(f"✅ Model Trained on {len(X_train)} days of history.")
print(f"   Tested on recent {len(X_test)} days.")
print("\n--- Accuracy Report ---")
print(classification_report(y_test, preds, target_names=['Bear', 'Sideways', 'Bull']))
```

```python
# ==========================================
# PART 4: DATE TESTING TOOL
# ==========================================

def test_market_regime(date_str):
    try:
        # Locate the date in our data
        target_date = pd.to_datetime(date_str)
        # Find nearest available trading day
        idx = df.index.get_indexer([target_date], method='nearest')[0]
        row = df.iloc[idx]
        actual_date = df.index[idx].date()

        # Get the 10 Parameters for that day
        input_data = row[features].values.reshape(1, -1)

        # Get AI Prediction
        prediction = rf_model.predict(input_data)[0]
        probs = rf_model.predict_proba(input_data)[0]

        regime_names = {0: "🐻 BEAR (Defensive)", 1: "🦀 SIDEWAYS (Hold)", 2: "🐂 BULL (Aggressive)"}

        print(f"\n==========================================")
        print(f"📅 DATE CHECK:     {actual_date}")
        print(f"💰 NIFTY PRICE:    {row['Close']:.2f}")
        print(f"==========================================")
        print(f"[Market Intelligence Inputs]")
        print(f"1. RSI (Momentum): {row['RSI']:.2f}  (>70 Overbought, <30 Oversold)")
        print(f"2. Volatility:     {row['Volatility']:.4f} (Higher = More Fear)")
        print(f"3. ADX (Trend):    {row['ADX']:.2f}   (>25 = Strong Trend)")
        print(f"4. SMA200 Dist:    {row['Dist_SMA200']*100:.2f}% (Dist from Long Term Trend)")
        print(f"...and 6 other parameters.")
        print(f"==========================================")
        print(f"🤖 AI PREDICTION:  {regime_names[prediction]}")
        print(f"📊 CONFIDENCE:     {max(probs)*100:.1f}%")
        print(f"==========================================")

    except Exception as e:
        print(f"Error: Data not found for {date_str}. (Check if date is within 2008-2024)")

# --- RUN YOUR TESTS BELOW ---

# TEST 1: The 2020 Covid Crash
test_market_regime('2020-03-23')

# TEST 2: The 2021 Bull Run
test_market_regime('2021-10-18')

# TEST 3: A Recent Day
test_market_regime('2024-02-01')

#TEST 4: Yesterday
test_market_regime('2025-11-27')
```
